### np.sort

Return a sorted copy of an array.

https://numpy.org/doc/stable/reference/generated/numpy.sort.html

```python
# code
import numpy as np
a = np.random.randint(1,100,15)
a
```

```python
b = np.random.randint(1,100,24).reshape(6,4)
b
```

```python
np.sort(a)
```

```python
# reverse [::-1]

np.sort(a)[::-1]
```

```python
# axis=1 : row wise sorting

np.sort(b, axis=1)
```

```python
# axis=0 : column wise sorting

np.sort(b,axis=0)
```

### np.concatenate

numpy.concatenate() function concatenate a sequence of arrays along an existing axis.

https://numpy.org/doc/stable/reference/generated/numpy.concatenate.html

```python
# code
c = np.arange(6).reshape(2,3)
d = np.arange(6,12).reshape(2,3)

print(c)
print()
print(d)
```

```python
np.concatenate((c,d), axis=0)   # axis=0 row wise concatenate
```

```python
np.concatenate((c,d), axis=1)   # axis=1 column wise concatenate
```

### np.where

The numpy.where() function returns the indices of elements in an input array where the given condition is satisfied.

https://numpy.org/doc/stable/reference/generated/numpy.where.html

```python
a
```

```python
# find all indices (index Position ) with value greater than 50
np.where(a>50)
```

```python
# replace all values > 50 with 0
np.where(a>50,False, True)
```

```python
# replace all values > 50 with 0
np.where(a>50,0,a)
```

```python
np.where(a%2 == 0, 0, a)
```
