```python
# write a function that can return the track record of 2 teams against each other

def func(team1, team2):
  print("Team 1 : ", team1)
  print("Team 2 : ", team2)
  df = ipl[((ipl['Team1'] == team2) & (ipl['Team2'] == team1)) | ((ipl['Team1'] == team1) & (ipl['Team2'] == team2)) ]
  count_matches = df.shape[0]
  team1_won_count = df[df['WinningTeam']== team1].shape[0]
  team2_won_count = df[df['WinningTeam']== team2].shape[0]
  print()
  print("{} numbers of matches Team 1 {} and Team 2 {} against each other".format(count_matches, team1, team2))
  print("Team 1 {} won {} matches against Team 2 {}".format(team1, team1_won_count, team2))
  print("Team 2 {} won {} matches against Team 1 {}".format(team2, team2_won_count, team1))

team1 = 'Royal Challengers Bangalore'
team2 = 'Mumbai Indians'
func(team1, team2)
```

```python
def func_1(team1, team2):
  # print("Team 1 : ", team1)
  # print("Team 2 : ", team2)
  df = ipl[((ipl['Team1'] == team2) & (ipl['Team2'] == team1)) | ((ipl['Team1'] == team1) & (ipl['Team2'] == team2)) ]
  count_matches = df.shape[0]
  team1_won_count = df[df['WinningTeam']== team1].shape[0]
  team2_won_count = df[df['WinningTeam']== team2].shape[0]
  # print()
  # print("{} numbers of matches Team 1 {} and Team 2 {} against each other".format(count_matches, team1, team2))
  # print("Team 1 {} won {} matches against Team 2 {}".format(team1, team1_won_count, team2))
  # print("Team 2 {} won {} matches against Team 1 {}".format(team2, team2_won_count, team1))
  return [team1, team2, count_matches, team1_won_count, team2_won_count]
```

```python
print(len(against_teams))
matches_won_data = []
for i in against_teams:
  team1, team2 = i[0], i[1]
  result = func_1(team1, team2)
  matches_won_data.append(result)
```

```python
matches_won_df = pd.DataFrame(matches_won_data, columns=['Team1','Team2','Against_matches_count','Team1_matches_won_count', 'Team2_matches_won_count'])
matches_won_df
```
