# Operators in Python

- Arithmetic Operators
- Relational Operators
- Logical Operators
- Bitwise Operators
- Assignment Operators
- Membership Operators

```python
# Arithmetric Operators
print(5+6)

print(5-6)

print(5*6)

print(5/2)

print(5//2)  # integer division

print(5%2) # modulus operator gives remainder

print(5**2)  # power 5 to power of 2
```

```python
# Relational Operators
print(4>5)  # greater than

print(4<5) # less than

print(4>=4) # greater than equal to

print(4<=4)  # less than equal to

print(4==4)  # equal to 

print(4!=4)  # not equal to
```

```python
# Logical Operators
print(1 and 0) # onle both 1 then return 1

print(1 or 0)  # any one is 1 then return 1

print(not 1)  # return opposite
```

```python
# Bitwise Operators

# bitwise and
print(2 & 3)

# bitwise or
print(2 | 3)

# bitwise xor ^  _ if in binary same bit ie 00 or  11 then return 0 otherwise 1 
# 010
# 011
# ---
# 001

print(2 ^ 3)

# negation
print(~3)

# left shift 
print(4 >> 2)

# right shift
print(5 << 2)
```

```python
# Assignment Operators

# = 
# a = 2

a = 2

# a = a % 2
a %= 2

# a++ ++a

print(a)
```

```python
# Membership Operators

# in/not in

print('D' not in 'Delhi')

print(1 in [2,3,4,5,6])
```

```python
# Program - Find the sum of a 3 digit number entered by the user

number = int(input('Enter a 3 digit number'))

# 345%10 -> 5
a = number%10

number = number//10

# 34%10 -> 4
b = number % 10

number = number//10
# 3 % 10 -> 3
c = number % 10

print(a + b + c)
```

```python
# Abhishek Code - using for loops for above problem for any number of digits
number = int(input('Enter any number for its digit sum'))

sum_dig = 0
while number > 0:
    sum_dig  = sum_dig + (number % 10)
    number = number // 10

print(sum_dig)
```

```python
# login program and indentation
# email -> sk@gmail.com
# password -> 1234

email = input('enter email')
password = input('enter password')

if email == 'sk@gmail.com' and password == '1234':
  print('Welcome')
elif email == 'sk@gmail.com' and password != '1234':
  # tell the user
  print('Incorrect password')
  password = input('enter password again')
  if password == '1234':
    print('Welcome,finally!')
  else:
    print('beta tumse na ho paayega!')
else:
  print('Not correct')
```

```python
# min of 3 number

a = int(input('first num'))
b = int(input('second num'))
c = int(input('third num'))

if a<b and a<c:
  print('smallest is',a)
elif b<c:
  print('smallest is',b)
else:
  print('smallest is',c)
```

```python
# menu driven calculator
menu = input("""
Hi! how can I help you.
1. Enter 1 for pin change
2. Enter 2 for balance check
3. Enter 3 for withdrawl
4. Enter 4 for exit
""")

if menu == '1':
  print('pin change')
elif menu == '2':
  print('balance')
elif menu == '3':
    print('withdrawl')
else:
  print('exit')
```

### Program - The current population of a town is 10000. The population of the town is increasing at the rate of 10% per year. You have to write a program to find out the population at the end of each of the last 10 years.

```python
# consider x is population of 9th year
# then 10th year population is 
# x + 10% of x = 10000
# x + 0.1x = 10000
# 1.1x = 10000
# x = 10000/1.1
```

```python
import math
curr_pop = 10000

for i in range(10,0,-1):
  print(i,curr_pop)
  # curr_pop = curr_pop - 0.1*curr_pop
  curr_pop = math.floor(curr_pop/1.1)
```

### Sequence sum

1/1! + 2/2! + 3/3! + ...

```python
import math
# code here
n = 3

s_sum = 0
for i in range(1, n+1):
    s_sum += i/math.factorial(i)

print(s_sum)
```
